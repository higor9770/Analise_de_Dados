```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Criando as listas que formarão um data frame**

```python
Ano = [2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
```

```python
TaxaDestribuicaoPessoa = [72.4, 75.9, 78.2, 79.5, 81.2, 80.9, 81.3, 78.1, 75.6, 70.6, 66.5, 59.0, 51.4, 46.7, 43.3]
```

```python
numeropessoa = [215.917, 228.543 , 237.860, 243.738, 251.088, 252.167, 255.207, 247.090, 240.993 , 226.819 , 214.881 ,	191.909 , 168.158 , 153.260 ,142.816 ]
```

**Criando o DataFrame**

```python
listas = list(zip(numeropessoa, TaxaDestribuicaoPessoa,))
table = pd.DataFrame(listas, index = Ano, columns = ['Número total de prescrições', 'Taxa de distribuição de opioides por 100 pessoas'])
table
```

**Tipos das colunas do DataFrame**

```python
table.dtypes
```

**Retirando a coluna do dataframe para aprender como fazer**

```python
outPrescricoes = table['Número total de prescrições']
# criando uma nova variável para armazenar os valores da coluna
```

```python
outDistribuicoes = table['Taxa de distribuição de opioides por 100 pessoas']
# criando uma nova variável para armazenar os valores da coluna
```

**Criando gráfico de linha para representar as duas variáveis**

```python
plt.rcParams["figure.figsize"] = (14, 8) # o tipo e tamanho do gráfico que será usado
plt.plot(Ano, outPrescricoes, color =  'darkgreen', marker = 'o') # inicializa as variaveis que serão plotadas
plt.title('Prescrições dispensadas de remédios contendo opióides', fontsize = 25) # o titulo que será usado
plt.xlabel('Ano', fontsize = 20) # nomeando o eixo horizontal
plt.ylabel('Prescrições', fontsize = 20) # nomeando o eixo vertical
plt.legend(['Mil'], loc = 'best', fontsize = 15) # legenda da linha do gráfico
plt.grid(True) # grade
plt.show() # para mostrar de modo mais estético (sempre por último)
```

```python
plt.rcParams["figure.figsize"] = (14, 8)
plt.bar(Ano, outDistribuicoes, color =  'darkgreen') # mostrando um grafico de barras (bar)
plt.title('Taxa de destribuição de prescrições por 100 pessoas', fontsize = 25)
plt.xlabel('Ano', fontsize = 20)
plt.ylabel('Distribuicoes', fontsize = 20)
plt.show()
```
