```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
empresas = ['Apple', 'Google', 'Amazon', 'Microsoft', 'tecent', "McDonald's", 'Visa', 'Facebook', 'Alibaba', 'Louis vuitton']
```

```python
bilhoes = [947.062, 819.573, 705.646, 611.460, 214.023, 196.526, 191.032,  186.421, 169.966, 124.273]
```

```python
plt.rcParams["figure.figsize"] = (15, 8) # o tipo e tamanho do gráfico que será usado
plt.plot(empresas, bilhoes, color =  'darkgreen', marker = 'o') # inicializa as variaveis que serão plotadas
# lw = aumentar o tamanho da linha  (colocar no plt.plot)
# marker = 'o' deixa pontilhado com os indices (colocar no plt.plot)
plt.title('Empresas mais ricas do mundo', fontsize = 25) # o titulo que será usado
plt.xlabel('Empresas', fontsize = 20) # nomeando o eixo horizontal
plt.ylabel('Bilhões', fontsize = 20) # nomeando o eixo vertical
plt.legend(['Bilhões de dolares'], loc = 'right', fontsize = 15) # legenda da linha do gráfico
plt.grid(True) # grade
plt.show() # para mostrar de modo mais estético (sempre por último)
```

```python
# mudando o tipo do gráfico

plt.rcParams["figure.figsize"] = (15, 8) # o tipo e tamanho do gráfico que será usado
plt.bar(empresas, bilhoes, color =  'darkgreen') # inicializa as variaveis que serão plotadas em barras
plt.title('Empresas mais ricas do mundo', fontsize = 25) # o titulo que será usado
plt.xlabel('Empresas', fontsize = 20) # nomeando o eixo horizontal
plt.ylabel('Bilhões', fontsize = 20) # nomeando o eixo vertical
plt.legend(['Bilhões de dolares'], loc = 'best', fontsize = 15) # legenda da linha do gráfico e localiza o lugar (canto superior direito)
plt.show() # para mostrar de modo mais estético (sempre por último)
```

Tipos de localizações das legendas

best - o melhor lugar (selecionado automaticamente)
upper right
upper left
lower left
lower right
right
center left
center right
lower center

```python
cores = ['black', 'dimgray', 'dimgrey', 'yellow', 'grey',
         'darkgray', 'darkgrey', 'silver', 'lightgray', 'lightgrey']
```

```python
plt.rcParams["figure.figsize"] = (15, 8) # o tipo e tamanho do gráfico que será usado
plt.bar(empresas, bilhoes, color = cores, edgecolor = 'black') # contorno da barra
plt.title('Empresas mais ricas do mundo', fontsize = 25) # o titulo que será usado
plt.xlabel('Empresas', fontsize = 20) # nomeando o eixo horizontal
plt.ylabel('Bilhões', fontsize = 20) # nomeando o eixo vertical
plt.show() # para mostrar de modo mais estético (sempre por último
```
